# 내 생일 그래프 그리기
## 201600282 엄기산
* 내가 태어난 년도부터 매년 내 생일의 최고 기온과 최저 기온의 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open('incheon_weather.csv','r')

data = csv.reader(f)

header = next(data)
print(header)
birth_year = []
birth_data1 = []
birth_data2 = []

for row in data:
    if (int(row[0].split('-')[0]) >= 1998) and (row[0].split('-')[1] == '01') and (row[0].split('-')[2] == '03') and (row[-1] != '') and (row[-2] != '') :
        birth_year.append(row[0].split('-')[0])
        birth_data1.append(float(row[-2]))
        birth_data2.append(float(row[-1]))
        
f.close()
        
        
```

```python
print(birth_data1)
print(birth_data2)
print(birth_year)
```

```python
plt.rcParams['axes.unicode_minus'] = False
plt.rc('font', family = 'Malgun Gothic')
plt.plot(birth_year,birth_data1, color = 'hotpink',ls='-.',marker='v',label ='최저')
plt.plot(birth_year,birth_data2, color = 'skyblue',ls=':',marker='>',label = '최고')
plt.title('98년 이후 1월 3일 최고 최저 기온')
plt.legend(loc=7)
plt.show()
```
