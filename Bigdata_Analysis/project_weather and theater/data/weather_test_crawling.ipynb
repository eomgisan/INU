```python
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from time import sleep

import csv


f = open('weather_test.csv','w', encoding='utf-8', newline='')
wr = csv.writer(f)
wr.writerow(["날짜", "날씨"])


url = "https://www.weatheri.co.kr/bygone/bygone01.php"

driver = webdriver.Chrome('../chromedriver.exe')
driver.implicitly_wait(1)
                          
driver.get(url)
                          
select_start = Select(driver.find_element_by_name("s_year"))
driver.implicitly_wait(1)

select_start.select_by_visible_text('2018')
driver.implicitly_wait(1)

select_start = Select(driver.find_element_by_name("s_month"))
driver.implicitly_wait(1)
select_start.select_by_value('1')
driver.implicitly_wait(1)

select_start = Select(driver.find_element_by_name("s_day"))
driver.implicitly_wait(1)
select_start.select_by_value('1')
driver.implicitly_wait(1)

select_end = Select(driver.find_element_by_name("e_year"))
driver.implicitly_wait(1)
select_end.select_by_value('2018')
driver.implicitly_wait(1)

select_end = Select(driver.find_element_by_name("e_month"))
driver.implicitly_wait(1)
select_end.select_by_value('12')
driver.implicitly_wait(1)

select_end = Select(driver.find_element_by_name("e_day"))
driver.implicitly_wait(1)
select_end.select_by_value('31')
driver.implicitly_wait(1)

driver.find_element_by_xpath("/html/body/table[2]/tbody/tr[3]/td[2]/table/tbody/tr[2]/td/img").click()
driver.implicitly_wait(1)

frame = driver.find_element_by_name("kako")
driver.switch_to.frame(frame)

for i in range(12):
    select = Select(driver.find_element_by_xpath("/html/body/table/tbody/tr/td/form/table[1]/tbody/tr/td[1]/select"))
    driver.implicitly_wait(1)
    select.select_by_index(i)
    driver.implicitly_wait(1)
    driver.find_element_by_xpath("/html/body/table/tbody/tr/td/form/table[1]/tbody/tr/td[2]/img").click()
    driver.implicitly_wait(1)
    
    table = driver.find_element_by_xpath('/html/body/table/tbody/tr/td/form/table[2]/tbody/tr/td/table')
    driver.implicitly_wait(1)
    tbody = table.find_element_by_xpath('/html/body/table/tbody/tr/td/form/table[2]/tbody/tr/td/table/tbody')
    driver.implicitly_wait(1)
    rows = tbody.find_elements_by_xpath('/html/body/table/tbody/tr/td/form/table[2]/tbody/tr/td/table/tbody/tr')
    driver.implicitly_wait(1)
    
    
    
    for index, value in enumerate(rows):
        if(index!=0):
            date = value.find_elements_by_tag_name("td")[0].text
            date = date.replace('\n',"")
            date = date.replace('년',"년 ")
            date = date.replace('월',"월 ")
            date = date.replace('일 ',"일")
            date = date.replace("(월 )","(월)")
            driver.implicitly_wait(1)
            
            weather = value.find_elements_by_tag_name("td")[10].text
            driver.implicitly_wait(1)
            
            wr.writerow([date, weather])
    print(i)
    sleep(5)
    

f.close()         
```
