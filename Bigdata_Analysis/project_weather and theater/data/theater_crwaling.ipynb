```python
from bs4 import BeautifulSoup

from selenium import webdriver
from selenium.webdriver.support.ui import Select
from time import sleep
from selenium.webdriver.common.action_chains import ActionChains
from datetime import datetime, timedelta
from selenium.webdriver.common.keys import Keys

import csv


f = open('theater.csv','a', encoding='utf-8', newline='')
wr = csv.writer(f)
wr.writerow(["날짜", "총 예매인원"])


url = "https://www.kobis.or.kr/kobis/business/stat/boxs/findDailyBoxOfficeList.do"

driver = webdriver.Chrome('../chromedriver.exe')

driver.implicitly_wait(1)
                          
driver.get(url)



select_lot = Select(driver.find_element_by_id("sWideAreaCd"))
driver.implicitly_wait(1)
select_lot.select_by_value('0105001')
driver.implicitly_wait(1)

start_search = "2016-04-12"
format = '%Y-%m-%d'
epoch = 0
while(start_search != '2018-01-01'):

    date_field = driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[3]/form/div/div[1]/div/div[2]/label")
    driver.implicitly_wait(1)



    ActionChains(driver).move_to_element(date_field).click().send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.DELETE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(Keys.BACK_SPACE).send_keys(start_search).perform()
    driver.implicitly_wait(1)



    driver.find_element_by_xpath("/html/body/div[1]/div[2]/div[2]/div[3]/form/div/div[5]/button").click()
    driver.implicitly_wait(1)
    

    dt_datetime = datetime.strptime(start_search,format)
    str_datetime = dt_datetime + timedelta(days=1)
    start_search = datetime.strftime(str_datetime,format)

    table = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[4]/div[3]/table')
    driver.implicitly_wait(1)
    tbody = table.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[4]/div[3]/table/tbody')
    driver.implicitly_wait(1)
    rows = tbody.find_elements_by_xpath('/html/body/div[1]/div[2]/div[2]/div[4]/div[3]/table/tbody/tr')
    driver.implicitly_wait(1)

    date = driver.find_element_by_xpath('/html/body/div[1]/div[2]/div[2]/div[4]/div[2]/h4').text
    driver.implicitly_wait(1)

    count = 0        
    for index, value in enumerate(rows):
        str_count = value.find_elements_by_tag_name("td")[7].text
        str_count = str_count.replace(',','')
        count += int(str_count)

    wr.writerow([date, count])

    sleep(1)
    epoch += 1
    if(epoch == 365):
        epoch = 0
        print("1년 단위 완료")

f.close()
```
